```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def MaclaurinSeries(x, n):

    # f(x) = cos(x)
    # g(0) = cos(x) and other higher order derivatives
    return ((-1)**n*(x**(2*n))/math.factorial(2*n))

def plot_MaclaurinSeries(x_range=(-2*np.pi, 2*np.pi), steps=10):

    x = np.linspace(x_range[0], x_range[1], 400)# Generate x values
    y_series = np.zeros_like(x, dtype=float) # initializing the series with 0
    y_actual = np.cos(x) # computing cos(x) for comparison

    for n in range(steps):

        y_series += np.float128(MaclaurinSeries(x, n)) # adding the Nth term to uptade y_series

        plt.figure(figsize=(6, 4))
        plt.plot(x, y_series, label=f'Maclaurin Series (steps={n*2})', linestyle='--')
        plt.plot(x, y_actual, label='cos(x)', color='red')
        plt.xlim(-10, 10)
        plt.ylim(-1.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Maclaurin Series Approximation of cos(x)')
        plt.legend()
        plt.show()

```

```python
plot_MaclaurinSeries(steps=10)
```

```python
def TaylorSeries(x, n):
    # f(x) = 1/x
    # g(1) = 1 - (x-1) + (x-1)^2 ...
    # Because we chose x = 1 factorials can be canceled and we get this formula of
    # sum (-1)^n(x-1)^n (factorials of alternating signs)
    return (-1)**n*((x-1)**n)

def plot_TaylorSeries(x_range=(-2*np.pi, 2*np.pi), steps=20):

    x = np.linspace(x_range[0], x_range[1], 400)
    y_series = np.zeros_like(x, dtype=float)
    y_actual = 1/x

    for n in range(steps):

        y_series += np.float128(TaylorSeries(x, n)) # adding the Nth term to uptade y_series

        plt.figure(figsize=(6, 4))
        plt.plot(x, y_series, label=f'Taylor Series (steps={n})', linestyle='--')
        plt.plot(x, y_actual, label='cos(x)', color='red')
        plt.xlim(-10, 10)
        plt.ylim(-60, 60)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Taylor Series Approximation of 1/x')
        plt.legend()
        plt.show()
```

```python
plot_TaylorSeries(steps=20)
# It's possible to note that the region of the function when x < 0 isn't describe at all;
# And it is also possible to see the tail of the function swinging around because of the alternating sign.
```
